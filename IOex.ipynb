```python
while True:
    try:
        userInput = int(input('1. Say hello / 2. Calculate square / 3. Exit'))

        if userInput == 1:
            print('Hello !')
        elif userInput == 2:
            try:
                userNum = int(input('Enter a number : '))
                print(f'The square of {userNum} is {userNum ** 2}')

            except ValueError:
                print('Invalid input. Please enter an integer.')
        elif userInput == 3:
            print('Exiting...')
            break

    except ValueError:
        print('Invalid input. Please enter 1, 2 or 3.')
```
