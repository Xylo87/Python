```python
def removeChars(word, n):
    if n >= len(word):
        print('Too many chars to remove')
    else:
        print(word[n::])

removeChars('Lycanthrope', 5)
```

```python
original_number = 1441

def isPalindrome(num):
    if num < 0:
        print('Your number must be positive')

    else:
        numList = list(str(num))
        numList.reverse()
        numJoin = int(''.join(numList))
        
        if num == numJoin:
            print(f'Yes, {original_number} is a palindrome')
        else:
            print(f'No, {original_number} is not a palindrome')

isPalindrome(original_number)
```

```python
list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]

def oneList(list1, list2):
    fList = []

    for x in list1:
        if x % 2 != 0:
            fList.append(x)

    for x in list2:
        if x % 2 == 0:
            fList.append(x)

    return fList

print(oneList(list1, list2))
```

```python
incomeEx = 27000

def incomeTax(income):
    totalTax = 0
    
    if income <= 10000:
        return totalTax
    
    elif income <= 20000:
        totalTax = (income - 10000) * 10/100
        return totalTax
    
    else:
        totalTax = ((income - 20000) * 20/100) + 1000
        return totalTax

incomeTax(incomeEx)
```

```python
def exponent(base, exp):
    if exp < 0:
        print('Exponent must be a non-negative integer')
    
    else:
        finalNum = base ** exp
        print(f'{base} to the power of {exp} is : {finalNum}')

exponent(5, 4)
```

```python
primeNums = []

for num in range(2, 21):
    est_premier = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            est_premier = False
            break
    if est_premier:
        primeNums.append(num)

for num in range(0, len(primeNums), 2):
    print(primeNums[num])
```

```python
userStr = input('Enter a string')

def containDigit(str):
   
    for char in str:
        if char.isdigit():
            return 'The string contains at least one digit.'
    
    return 'The string does not contain any digits.'

check = containDigit(userStr)
print(check)
```

```python
def capStr(str):
    newStr = ''

    for word in str.split():
        
        capWord = word.capitalize()
        newStr += f'{capWord} '

    return newStr

userStr = 'Toubon est un beau gosse de fou furieux '

userStrCap = capStr(userStr)
print(userStrCap)
```

```python
import time

def countDown(count):

    while count > 0:
        print(f'Time remaining: {count} seconds')
        time.sleep(1)
        count -= 1
    
    print('Time\'s up !')

timeCount = 5
countDown(timeCount)
```
