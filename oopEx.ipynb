```python
class Vehicle:
    # color = "White"

    def __init__(self, name, max_speed, milenage, color='White'):
        self.name = name
        self.max_speed = max_speed
        self.milenage = milenage
        self.color = color

    def seating_capacity(self, capacity):
        return f'The seating capacity of a {self.name} is {capacity} passengers'

honda = Vehicle('Civic', 240, 14000)
print(honda.name)
print(honda.max_speed)
print(honda.milenage)
```

```python
class VehicleEx:
    def __init__(self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100
    
class BusEx(VehicleEx):
    def fare(self):
        finalAmount = super().fare() + (self.capacity / 10 * 100)
        return finalAmount

busEx = BusEx('LongLine', 50000, 50)
print(busEx.fare())
```

```python
print(issubclass(BusEx, VehicleEx))
```

```python
from math import pi

class Shape:
    def area(self):
        raise NotImplementedError("Area method must be implemented by subclasses")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return round((self.radius**2)*pi, 5)

class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side**2

shapes = [Circle(5), Square(7), Circle(3)]

for shape in shapes:
    print(shape.area())
```
