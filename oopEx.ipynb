```python
class Vehicule:
    def __init__(self, name, max_speed, milenage):
        self.name = name
        self.max_speed = max_speed
        self.milenage = milenage

honda = Vehicule('Civic', 240, 14000)
print(honda.name)
print(honda.max_speed)
print(honda.milenage)
```
