```python
import re

str1 = "P@#yn26at^&i5ve"
alphas = 0
numbers = 0
specs = 0

for char in str1:

    if char.isdigit():
        numbers += 1
    elif char.isalpha():
        alphas += 1
    elif re.search(r'[^a-zA-Z0-9]', char):
    # elif re.search(r'\W', char):
        specs += 1


print(f'Alphas : {alphas} \nNumbers : {numbers} \nSpecs : {specs}')
```

```python
s1 = "Abc"
s2 = "Xyz"

length = len(s1) if len(s1) > len(s2) else len(s2)

s2 = s2[::-1]

def mixedStr(str1, str2):
    s3 = ''

    for x in range(length):
        if x < len(str1):
            s3 += str1[x]

        if x < len(str2):
            s3 += str2[x]
    
    return s3

mixedStr(s1, s2)
```
