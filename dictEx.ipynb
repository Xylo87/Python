```python
my_dict = {'apple': 3, 'zebra': 1, 'banana': 2, 'cat': 4}

sortedKeys = sorted(my_dict.keys())

sortedDict = {k: my_dict[k] for k in sortedKeys}
print(sortedDict)
```

```python
my_dict = {'Jessa': 3, 'Kelly': 1, 'Jon': 2, 'Kerry': 4, 'Joy': 1}

sortedByValues = sorted(my_dict, key=lambda k: (my_dict[k], k))
print(sortedByValues)

sortedDict = {k: my_dict[k] for k in sortedByValues}
print(sortedDict)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'x': 10, 'y': 20, 'z': 10}

def isUnique(dictionnary):

    nbDictValues = len(dictionnary.values())
    nbSetValues = len(set(dictionnary.values()))

    if nbDictValues == nbSetValues:
        return True
    else:
        return False

print(isUnique(dict1))
```
