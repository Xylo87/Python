```python
import re

def countWords(file):

    try:
        with open(file, 'r') as sample:
            content = sample.read().lower()
            words = re.findall(r"\b[\w'-]+\b", content)
            return len(words)
        
    except FileNotFoundError:
        return "File doesn't exist"

countWords('sample.txt')
```

```python
def countChar(file):

    try:
        with open(file, 'r') as sample:
            content = sample.read()
            return len(content)
    
    except FileNotFoundError:
        return "File doesn't exist"
    
countChar('sample.txt')
```

```python
def countSpecWord(file, word):
    try:
        with open(file, 'r') as sample:
            content = sample.read().lower()
            words = re.findall(r"\b[\w'-]+\b", content)
            return words.count(word)

    except FileNotFoundError:
        return "File doesn't exist"

countSpecWord('sample.txt', 'et')
```

```python
def fileCopy(inputFile, outputFile):
    try:
        with open(inputFile, 'r') as inputF:
            content = inputF.read()
    
        with open(outputFile, 'r+') as outputF:
            outputF.write(content)
            outputF.seek(0)
            newContent = outputF.read()
            print(newContent)
    
    except FileNotFoundError:
        return "One or several file(s) does not exist"
    
fileCopy('inputFile.txt', 'outputFile.txt')
```

```python
def renameFile(oldname, newName):
    try:
        os.rename(oldname, newName)
    except FileNotFoundError:
        print("File doesn't exist")

renameFile('renaming.txt', 'renaming2.txt')
```

```python
def deletingFile(file):
    if not os.path.exists(file):
        print("File doesn't exist")

    os.remove(file)

deletingFile('forDelete.txt')
```
