```python
import json

data = {"key1" : "value1", "key2" : "value2"}

toJSON = json.dumps(data)
print(toJSON)
```

```python
toParse = json.loads(toJSON)
print(toParse['key2'])
```

```python
sampleJson = {"key1": "value1", "key2": "value2"}

toJSONsample = json.dumps(sampleJson, indent=2, separators=(",", " = "))
print(toJSONsample)
```

```python
sampleJson2 = {"id" : 1, "name" : "value2", "age" : 29}

with open('sample.json', 'w') as outfile:
    json.dump(sampleJson2, outfile, indent=2, sort_keys=True)
```
