```python
def calculation(a, b):
    calcSum = a + b
    calcSub = a - b

    return calcSum, calcSub

calculation(89, 76)
```

```python
def show_employees(name, salary = 9000):
    print(f'Name : {name}')
    print(f'Salary: {salary}')

show_employees('Thomas')
```

```python
def outer(a, b):

    def inner():
        sum = a + b
        return sum
    
    finalSum = inner()

    return finalSum

print(outer(10, 18))
```

```python
start = 10

def recurs(n):
    if n == 0:
        return 0
    else:
        return n + recurs(n - 1)
    
recurs(start)
```

```python
def evenNum():
    list = []

    for x in range(4, 30):
        if x % 2 == 0:
            list.append(x)

    return list

evenNum()
```

```python
x = [4, 6, 8, 24, 12, 2]

def findLarg(userList):
    currentNum = userList[0]

    for num in userList:
        if num > currentNum:
            currentNum = num
    
    return currentNum

findLarg(x)
```

```python
def describe_pet(animal_type, pet_name):
    print(f'Type : {animal_type}')
    print(f'Name : {pet_name}')

describe_pet('Dog', 'Wolfy')
describe_pet(animal_type = 'Dog', pet_name = 'Wolfy')
```

```python
def print_info(**keyargs):

    for key, value in keyargs.items():
        print(f'{key} = {value}')

print_info(Name = 'Isimorn', Job = 'Nuclear Engineer', Address = 'Villeurbanne')
```

```python
start = 19

def recurs(n):
    if n < 0:
        print('Your number must be a non-negative number')
    elif n == 0:
        return 1
    else:
        return n * recurs(n - 1)
    
recurs(start)
```
