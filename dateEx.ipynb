```python
from datetime import datetime

print(datetime.now())
# print(datetime.now().time())
```

```python
date_string = "Feb 25 2020 4:20PM"
dateFormat = '%b %d %Y %I:%M%p'

toObj = datetime.strptime(date_string, dateFormat)
print(toObj)
```

```python
from datetime import timedelta

given_date = datetime(2020, 2, 25)

subWeek = given_date - timedelta(weeks=1)
print(subWeek)
```

```python
given_date = datetime(2020, 2, 25)
dateFormat = '%A %d %B %Y'

toStr = datetime.strftime(given_date, dateFormat)
print(toStr)
```

```python
given_date = datetime(2020, 7, 26)

toStr = datetime.strftime(given_date, '%A')
print(toStr)
```

```python
given_date = datetime(2020, 3, 22, 10, 0, 0)

addWeekHours = given_date + timedelta(weeks=1, hours=12)
print(addWeekHours)
```

```python
print(int(datetime.now().timestamp() * 1000))
```

```python
given_date = datetime(2020, 2, 25)
dateFormat = '%Y-%m-%d %H:%M:%S'

toStr = datetime.strftime(given_date, dateFormat)
print(toStr)
```

```python
from dateutil.relativedelta import relativedelta

given_date = datetime.now().date()
fourMonths = given_date + relativedelta(months=4)

print(fourMonths)
```

```python
date_1 = datetime(2020, 2, 25)
date_2 = datetime(2020, 9, 17)

newDate = date_1 - date_2
print(f'{abs(newDate.days)} days')
```
