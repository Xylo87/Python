```python
number = 75869564212

def digitsCount(num):
    numToList = list(str(num))

    acc = 0
    while acc < len(numToList):
        acc += 1

    return acc

digitsCount(number)
```

```python
primeNums = []

for num in range(25, 50):
    est_premier = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            est_premier = False
            break
    if est_premier:
        primeNums.append(num)

for num in primeNums:
    print(num)
```

```python
def factorial(num):
    result = 1

    for x in range(1, num + 1):
        result *= x
    
    return result

factorial(5)
```

```python
def reverseInt(num):
    toList = reversed(list(str(num)))
    
    rev = ''.join(toList)

    return int(rev)

reverseInt(76542)
```

```python
num = 2
terms = 5

def sumSeries(num, terms):
    acc = 0

    for x in range(1, terms + 1):
        currentNum = int(f'{num}' * x)
        acc += currentNum
    
    return acc

sumSeries(2, 5)
```

```python
nested_list = [1, [2, 3], [4, 5, 6], 7, [8, 9]]
flatList = []

def flat(userList):
    for el in userList:

        if isinstance(el, list):
            print('ok')
            for e in el:
                flatList.append(e)
        else:
            flatList.append(el)

    return flatList
    
flat(nested_list)
```

```python
num1 = 9876543210
num2 = -5082

def highLow(number):
    toList = list(str(abs(number)))
    currentHigh = int(toList[0])
    currentLow = int(toList[0])

    for el in toList:
        if int(el) > currentHigh:
            currentHigh = int(el)
    
    for el in toList:
        if int(el) < currentLow:
            currentLow = int(el)
    
    return [currentHigh, currentLow]

largeSmall = highLow(num2)

print(f'Largest digit in {num2}: {largeSmall[0]}')
print(f'Smallest digit in {num2}: {largeSmall[1]}')
```
